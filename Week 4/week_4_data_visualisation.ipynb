```python
# import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load datasets

mat = pd.read_csv("student-mat.csv")
por = pd.read_csv("student-por.csv")

print("Math Dataset Shape:", mat.shape)
print("Portuguese Dataset Shape:", por.shape)
```

```python
# quick preview of the dataset
print("Math dataset sample:")
display(mat.head())

print("\nPortuguese dataset sample:")
display(por.head())
```

```python
# dataset info
print("\n--- Maths info ---")
mat.info()
```

```python
print("\n--- Portuguese info ---")
por.info()
```

```python
# basic statistics for grades and numerical columns

print("Maths grades (G1,G2,G3) summary:")
display(mat[['G1','G2','G3']].describe().T)
```

```python
print("\nPortuguese grades (G1,G2,G3) summary:")
display(por[['G1','G2','G3']].describe().T)
```

```python
# handling missing values and cleaning if any exists

print("Missing values in Maths dataset:")
display(mat.isnull().sum()[lambda s: s>0])
```

```python
print("\nMissing values in Portuguese dataset:")
display(por.isnull().sum()[lambda s: s>0])
```

```python
# combining datasets for joint analysis

mat_copy = mat.copy()
por_copy = por.copy()
mat_copy['subject'] = 'Maths'
por_copy['subject'] = 'Portuguese'

combined = pd.concat([mat_copy, por_copy], ignore_index=True)
print("Combined Dataset Shape:", combined.shape)
```

```python
# a histogram to visualize the distribution of final grades (G3) for both subjects

plt.figure(figsize=(10,6))
plt.hist(mat['G3'], bins=20, alpha=0.6, label='Maths (G3)', edgecolor='black')
plt.hist(por['G3'], bins=20, alpha=0.6, label='Portuguese (G3)', edgecolor='black')
plt.title("Distribution of Final Grades (G3) — Maths vs Portuguese")
plt.xlabel("Final Grade (0-20)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# a bar chart to look at the Average final grade (G3) by weekly studytime for Maths

avg_by_study_maths = mat.groupby('studytime', as_index=False)['G3'].mean().sort_values('studytime')

plt.figure(figsize=(8,5))
plt.bar(avg_by_study_maths['studytime'].astype(str), avg_by_study_maths['G3'])
plt.title("Average Maths Final Grade (G3) by Weekly Study Time")
plt.xlabel("Study Time Category (1 = <2h, 2 = 2-5h, 3 = 5-10h, 4 = >10h)")
plt.ylabel("Average Final Grade (G3)")
plt.tight_layout()
plt.show()
```

```python
# another bar chart to look at the average final grade (G3) by weekly studytime for Portuguese

avg_by_study_portuguese = por.groupby('studytime', as_index=False)['G3'].mean().sort_values('studytime')

plt.figure(figsize=(8,5))
plt.bar(avg_by_study_portuguese['studytime'].astype(str), avg_by_study_portuguese['G3'])
plt.title("Average Portuguese Final Grade (G3) by Weekly Study Time")
plt.xlabel("Study Time Category (1 = <2h, 2 = 2-5h, 3 = 5-10h, 4 = >10h)")
plt.ylabel("Average Final Grade (G3)")
plt.tight_layout()
plt.show()
```

```python
# a line chart to look at the average grade progression G1 -> G2 -> G3 in Maths

means = mat[['G1','G2','G3']].mean()
plt.figure(figsize=(8,5))
plt.plot(['G1','G2','G3'], means.values, marker='o')
plt.title("Average Grade Progression (Maths)")
plt.xlabel("Period")
plt.ylabel("Average Grade")
plt.ylim(0, 20)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# another line chart to look at the average grade progression G1 -> G2 -> G3 in Portuguese

means = por[['G1','G2','G3']].mean()
plt.figure(figsize=(8,5))
plt.plot(['G1','G2','G3'], means.values, marker='o')
plt.title("Average Grade Progression (Portuguese)")
plt.xlabel("Period")
plt.ylabel("Average Grade")
plt.ylim(0, 20)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# we can further visualise to see the average final grade by sex (Male of Female) in both subjects

avg_by_sex_subject = combined.groupby(['subject','sex'], as_index=False)['G3'].mean().pivot(index='sex', columns='subject', values='G3')
avg_by_sex_subject = avg_by_sex_subject[['Maths','Portuguese']]  # ensure order

# here we can plot side-by-side bars for good comparison

labels = avg_by_sex_subject.index.astype(str).tolist()
x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(7,5))
plt.bar(x - width/2, avg_by_sex_subject['Maths'], width=width, label='Maths')
plt.bar(x + width/2, avg_by_sex_subject['Portuguese'], width=width, label='Portuguese')
plt.xticks(x, labels)
plt.xlabel("Sex (F/M)")
plt.ylabel("Average Final Grade (G3)")
plt.title("Average Final Grade by Sex and Subject")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# furthermore we can plot a scatter to visualize the studytime vs final grade in Maths to show relationship

plt.figure(figsize=(8,5))
plt.scatter(mat['studytime'] + (np.random.rand(len(mat))-0.5)*0.15, mat['G3'], alpha=0.6, s=20)  # jitter x for readability
plt.xticks([1,2,3,4])
plt.xlabel("Study Time Category (1..4)")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade (Maths)")
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()

# Show correlation number
corr = mat['studytime'].corr(mat['G3'])
print(f"Pearson correlation (studytime vs G3) for Maths: {corr:.3f}")
```

```python
# one can be made for Portuguese as well

plt.figure(figsize=(8,5))
plt.scatter(por['studytime'] + (np.random.rand(len(por))-0.5)*0.15, por['G3'], alpha=0.6, s=20)  # jitter x for readability
plt.xticks([1,2,3,4])
plt.xlabel("Study Time Category (1..4)")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade (Portuguese)")
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()

# Show correlation number
corr = por['studytime'].corr(por['G3'])
print(f"Pearson correlation (studytime vs G3) for Portuguese: {corr:.3f}")
```

```python
# we can save a short textual report for a total summary

total_students_maths = mat.shape[0]
total_students_por = por.shape[0]
overall_avg_maths = mat['G3'].mean()
overall_avg_por = por['G3'].mean()

report = f"""

Maths dataset:
- Students: {total_students_maths}
- Average final grade (G3): {overall_avg_maths:.2f}

Portuguese dataset:
- Students: {total_students_por}
- Average final grade (G3): {overall_avg_por:.2f}

Key observations:
- Average grade progression (G1 -> G2 -> G3) indicates whether students improve over time.
- Higher studytime categories correspond to higher average final grades (positive correlation shown by scatter and correlation coefficient).
"""
```

```python
print(report)
```
